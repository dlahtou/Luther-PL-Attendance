```python
import pandas as pd
```

```python
with open('PLmatches.csv', 'r') as open_file:
    df = pd.read_csv(open_file, index_col=0)
```

```python
def get_weeks_from_end(row):
    return abs(row['matchweek'] - 11)
```

```python
df['week_power'] = df.apply(get_weeks_from_end, axis=1)
```

```python
df.drop(['weeks_from_end'], axis=1, inplace=True)
```

```python
with open('PLmatches.csv', 'w') as open_file:
    df.to_csv(open_file)
```
