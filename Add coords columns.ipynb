```python
import pandas as pd
import pickle as pkl
import re
```

```python
with open('PLmatches.csv', 'r') as open_file:
    df = pd.read_csv(open_file, index_col=0)
```

```python
with open('stadium_info.pkl', 'rb') as open_file:
    stadium_info_dict = pkl.load(open_file)
```

```python
for stadium in stadium_info_dict.keys():
    if not stadium_info_dict[stadium]['coords']:
        print('skipped {}'.format(stadium))
        continue
    if stadium == 'Liberty Stadium, Swansea':
        continue
    print(stadium)
    stadium_info_dict[stadium]['coords'] = re.search('..°.*′.*″N .°.*′.*″.',
                                                     stadium_info_dict[stadium]['coords']).group(0)
```

```python
stadium_info_dict['Liberty Stadium, Swansea']['coords']
```

```python
for stadium in stadium_info_dict.keys():
    if not stadium_info_dict[stadium]['coords']:
        continue
    if stadium == 'Liberty Stadium, Swansea':
        continue
    coords_str = stadium_info_dict[stadium]['coords']
    coords_both = coords_str.split(' ')
    
    Ncoords = coords_both[0]
    Wcoords = coords_both[1]
    
    wmult = 1
    if Wcoords[-1] == 'W':
        wmult = -1
    
    Ndeg = float(re.match('.*°', Ncoords).group(0)[:-1])
    Nmin = float(re.search('°.*′', Ncoords).group(0)[1:-1])/60
    Nsec = float(re.search('′.*″', Ncoords).group(0)[1:-1])/3600
    Ndec = (Ndeg + Nmin + Nsec)
    
    Wdeg = float(re.match('.*°', Wcoords).group(0)[:-1])
    Wmin = float(re.search('°.*′', Wcoords).group(0)[1:-1])/60
    Wsec = float(re.search('′.*″', Wcoords).group(0)[1:-1])/3600
    Wdec = wmult * (Wdeg + Wmin + Wsec)
    
    stadium_info_dict[stadium]['Ndec'] = Ndec
    stadium_info_dict[stadium]['Wdec'] = Wdec
```

```python
for stadium in stadium_info_dict.keys():
    stadium_info_dict[stadium]['Ndec'] = round(stadium_info_dict[stadium]['Ndec'], 4)
    stadium_info_dict[stadium]['Wdec'] = round(stadium_info_dict[stadium]['Wdec'], 4)
```

```python
def set_Ndec_column(row):
    return stadium_info_dict[row['match_stadium']]['Ndec']

def set_Wdec_column(row):
    return stadium_info_dict[row['match_stadium']]['Wdec']
```

```python
df['Ndec'] = df.apply(set_Ndec_column, axis=1)
```

```python
df['Wdec'] = df.apply(set_Wdec_column, axis=1)
```

```python
team_stadium_tupleset = set()
for index in df.index:
    team_stadium_tupleset.add((df.loc[index, 'hometeam_name'], df.loc[index, 'match_stadium']))
```

```python
team_stadium_tupleset
```

```python
team_stadium_dict = dict()
for tuple_ in team_stadium_tupleset:
    team_stadium_dict[tuple_[0]] = tuple_[1]
```

```python
def set_away_Ndec_column(row):
    return stadium_info_dict[team_stadium_dict[row['awayteam_name']]]['Ndec']

def set_away_Wdec_column(row):
    return stadium_info_dict[team_stadium_dict[row['awayteam_name']]]['Wdec']
```

```python
df['away_Ndec'] = df.apply(set_away_Ndec_column, axis=1)
df['away_Wdec'] = df.apply(set_away_Wdec_column, axis=1)
```

```python
df[:5]
```

```python
from geopy.distance import distance
```

```python
def get_stadium_distance(row):
    home_coords = (row['Ndec'], row['Wdec'])
    away_coords = (row['away_Ndec'], row['away_Wdec'])
    
    return float(distance(home_coords, away_coords).km)

float(get_stadium_distance(df.iloc[0,:]))
```

```python
df['away_distance'] = df.apply(get_stadium_distance, axis=1)
```
