```python
import pandas as pd
import numpy as np
from math import isnan
```

```python
with open('PLmatches.csv') as open_file:
    df = pd.read_csv(open_file, index_col=0)
```

```python
def get_prior_matches_by_stadium(match_stadium, season_num, match_timestamp, num_matches=38):
    all_prior_matches = df[(df['match_stadium'] == match_stadium)
                           &(df['season_num'] == season_num)
                          &(df['match_timestamp'] < match_timestamp)].sort_values(by='match_timestamp')
    
    return all_prior_matches[-num_matches:]

def get_mean_value(data, col_name):
    values = []
    for row_index in data.index:
        values.append(data.loc[row_index, col_name])
    
    return np.mean(values)
```

```python
def get_stadium_mean_attendance_delta(row):
    match_stadium = row['match_stadium']
    season_num = row['season_num']
    cur_timestamp = row['match_timestamp']
    
    df_lastgame = get_prior_matches_by_stadium(match_stadium, season_num, cur_timestamp)
    
    cur_attendance = row['match_attendance']
    mean_attendance = np.mean(get_mean_value(df_lastgame, 'match_attendance'))
    
    return cur_attendance - mean_attendance
```

```python
df['mean_attendance_delta'] = df.apply(get_stadium_mean_attendance_delta, axis=1)
```

```python
df.dropna(subset=['mean_attendance_delta'])
```

```python
with open('PLmatches.csv', 'w') as open_file:
    df.to_csv(open_file)
```
