```python
for file in sorted(listdir('weatherdata')):
    with open(f'weatherdata/{file}') as open_file:
        weather = json.load(open_file)
        try:
            accum = weather['daily']['data'][0]['precipAccumulation']
        except KeyError:
            accum = 0
        try:
            prob = weather['daily']['data'][0]['precipProbability']
        except KeyError:
            prob = 0
        try:
            hightemp = weather['daily']['data'][0]['temperatureHigh']
        except KeyError:
            pass
        
        print(f'{file}: {accum} // {prob} // {hightemp}')
```
