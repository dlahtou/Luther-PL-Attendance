```python
import pandas as pd
import requests
import datetime
import json
import numpy as np
```

```python
with open('PLmatches.csv', 'r') as open_file:
    df = pd.read_csv(open_file, index_col=0)
```

```python
df_ars = df[(df['match_stadium'] == 'Emirates Stadium, London')]
```

```python
df_ars = df_ars.dropna(subset=['match_attendance'])
```

```python
testy = df_ars.index
```

```python
def get_row_weather(row):
    if row['hometeam_name'] != 'Tottenham Hotspur' and row['hometeam_name'] != 'Arsenal':
        return pd.Series({"precip_prob": np.nan, "temp_high": np.nan})
    match_time = datetime.datetime.utcfromtimestamp(row['match_timestamp']//1000)
    match_midnight = match_time.replace(hour=0, minute=0, second=0)
    match_midnight = match_midnight + datetime.timedelta(days=1)
    mdnt_ts = int(match_midnight.timestamp())
    Ndec = str(row['Ndec'])
    Wdec = str(row['Wdec'])

    if isfile(f'weatherdata/{mdnt_ts}.json'):
        print(f'opening {mdnt_ts} from file')
        with open(f'weatherdata/{mdnt_ts}.json') as open_file:
            weather = json.load(open_file)
    else:
        test = requests.get(f'https://api.darksky.net/forecast/1c0e0cc4529f6cafd3ab0ef69720cb2a/{Ndec},{Wdec},{mdnt_ts}')
        weather = test.json()

        with open(f'weatherdata/{mdnt_ts}.json', 'w') as open_file:
            json.dump(weather, open_file)
    try:
        precip_prob = weather['daily']['data'][0]['precipProbability']
    except KeyError:
        precip_prob = np.nan
    try:
        temp_high = weather['daily']['data'][0]['temperatureHigh']
    except KeyError:
        temp_high = np.nan
        
    return pd.Series({"precip_prob": precip_prob, "temp_high": temp_high})
```

```python
df[['precip_prob', 'temp_high']] = df.apply(get_row_weather, axis=1)
```
